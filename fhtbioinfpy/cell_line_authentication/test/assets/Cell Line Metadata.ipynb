```python
# Files

# In-house RNA-seq dataset
query_expr_file = "./input_data/2022-02-10_heatmap_log2_TPM_r113153x1115.gctx"

# Read the sample_info containing the DepMap_ID and the cell line name
sample_info_file = "./input_data/sample_info.csv" 

# Read the metadata from Lab Guru to comapr sample and real cell line name
metadata_file = "./input_data/H202SC21053155 2021-05-25 next gen sequencing experiment annotations template - DS 5-25-2021_36 samples.csv"
```

## Read in-house RNA-seq dataset experiment

```python
# Read in-house RNA seq dataset 

query_expr_gctx = parse.parse(query_expr_file)
query_expr = query_expr_gctx.data_df

print(query_expr.shape)
query_expr.head()
```

```python
# Remove ENSMUSG (mouse) genes
query_expr = query_expr[query_expr.index.str.contains("ENSMUSG")==False]

print(query_expr.shape)
query_expr.head()
```

```python
# Check for NAs in index
print('NAs in the indexes: ', collections.Counter(list(query_expr.index.isna())))

# Check for NAs in columns
print('NAs in the columns: ', collections.Counter(list(query_expr.columns.isna())))

# Check for NaN in DataFrame
print('NAs in DataFrame: ', query_expr.isnull().sum().sum())

# View columns with all NAs (Mouse samples) if any
NA_col_query_expr = collections.Counter(query_expr.isnull().all())
if NA_col_query_expr[True] ==0:
    print("There is no columns with only NAs in the Dataframe.")
else:
    print('There is', NA_col_query_expr[True] , 'columns with only NAs in this Dataframe.')
    NA_samples = list(query_expr[query_expr.columns[query_expr.isnull().all()]].columns)
    print('Names of columns/samples with all NAs in DataFrame: \n', NA_samples)
    # Drop columns with all NAs in query_expr
    query_expr = query_expr.dropna(axis='columns')

# Sanity check
print('New # of NAs in DataFrame: ', query_expr.isnull().sum().sum())
```

```python
# Save file to csv
query_expr.to_csv('input_data/query_expr.csv')
```

## Add cell line name to the sample name

```python
# Read the metadata from 10_heatmap_log2_TPM_r113153x1115.gctx to compare the cell line sample 
# and the cell line with the highest correlation

metadata = pd.DataFrame(query_expr_gctx.col_metadata_df['bio_context_id'])
metadata.index.name = "sample"
metadata.drop(NA_samples, inplace= True) # Drop the samples with all NAs samples (Mouse samples)
metadata.reset_index(inplace=True)

print(metadata.shape)
metadata.head()
```

```python
# Check for NAs in index
print('NAs in the indexes: ', collections.Counter(list(metadata.index.isna())))

# Check for NAs in columns
print('NAs in the columns: ', collections.Counter(list(metadata.columns.isna())))

# Check for NaN in DataFrame
print('NAs in DataFrame: ', metadata.isnull().sum().sum())

# View samples with no annotation if any
sample_wth_no_annot = [sample for sample, row in metadata.iterrows() if row.isnull().any()]
if sample_wth_no_annot==[]:
    print('All samples have an annotation.')
else:
    print('Samples with no annotation: \n', sample_wth_no_annot)
    # Replace samples with no annotation by 'NoAnnotation' in metadata if NAs
    metadata.fillna('NOANNOTATION', inplace=True)

# Sanity Check
print('New # of NAs in DataFrame: ', metadata.isnull().sum().sum())
print('Dim of the DataFrame: ', metadata.shape)
```

## Add "cleaned" cell line name 

```python
metadata["cleaned_bio_context_id"] = metadata["bio_context_id"] 
# 1) force all names to be uppercase 
metadata["cleaned_bio_context_id"] = metadata["cleaned_bio_context_id"].str.upper()
# 2) remove all special characters before comparing
metadata["cleaned_bio_context_id"] = metadata["cleaned_bio_context_id"].str.replace('[^A-Za-z0-9]+','')
# 3) remove LabGuru_cleaned_annot_cl
metadata["cleaned_bio_context_id"] = metadata["cleaned_bio_context_id"].str.replace('NCI','')

print(metadata.shape)
metadata.head()
```

## Add DepMap_ID

```python
sample_info = pd.read_csv(sample_info_file)
# sample_info.set_index(['DepMap_ID'], inplace=True)
sample_info = sample_info[['DepMap_ID', 'stripped_cell_line_name']]

print(sample_info.shape)
sample_info.head()
```

```python
# Check for NAs in index
print('NAs in the DepMap_ID indexes: ', collections.Counter(list(sample_info.index.isna())))

# Check for NAs in columns
print('NAs in the columns: ', collections.Counter(list(sample_info.columns.isna())))

# Check for NaN in DataFrame
print('NAs in DataFrame: ', sample_info.isnull().sum().sum())

# Subset rows with NAs
print('Row(s) with NA: \n', sample_info[sample_info.isnull().any(axis=1)])

# Drop NAs
sample_info.dropna(subset = ["stripped_cell_line_name"], inplace=True)

# Sanity Check
print('New # of NAs in DataFrame: ', sample_info.isnull().sum().sum())
print('Dim of the DataFrame: ',sample_info.shape)
```

```python
sample_info["cleaned_stripped_cell_line_name"] = sample_info["stripped_cell_line_name"] 

# 1) force all names to be uppercase 
sample_info["cleaned_stripped_cell_line_name"] = sample_info["cleaned_stripped_cell_line_name"].str.upper()
# 2) remove all special characters before comparing
sample_info["cleaned_stripped_cell_line_name"] = sample_info["cleaned_stripped_cell_line_name"].str.replace('[^A-Za-z0-9]+','')
# 3) remove NCI
sample_info["cleaned_stripped_cell_line_name"] = sample_info["cleaned_stripped_cell_line_name"].str.replace('NCI','')

print(sample_info.shape)
sample_info.head()
```

```python
sample_df = metadata.merge(sample_info, how='left', left_on='cleaned_bio_context_id', right_on='cleaned_stripped_cell_line_name').drop('cleaned_stripped_cell_line_name',1)

print(sample_df.shape)
sample_df.tail()
```
